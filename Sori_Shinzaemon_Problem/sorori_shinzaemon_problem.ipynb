## 3. Compute Daily and Total Grains (for 100 days)
This assignment requires computing:
- The number of grains received **each day** for 100 days
- The **cumulative total** up to each day

The implementation below uses **for loops** only (no exponentiation).

```python
# Compute daily grains and cumulative totals for n_days = 100 using for-loops
n_days = 100
list_n_grains = []         # grains per day
list_total_grains = []     # cumulative grains
total = 0

grains_today = 1  # Day 1
for day in range(1, n_days + 1):
    if day == 1:
        grains_today = 1
    else:
        grains_today *= 2  # double each day

    list_n_grains.append(grains_today)
    total += grains_today
    list_total_grains.append(total)

print("Rice grains on day 100:", list_n_grains[-1])
print("Total rice grains after 100 days:", list_total_grains[-1])
```


## 5. Parameterization with Functions
This assignment requires a function that accepts the number of days as an argument and returns two lists:
- Grains per day
- Cumulative grains

The function below uses **for-loops** for clarity.


```python

def compute_sorori_shinzaemon(n_days=100):
    """
    Compute the daily and cumulative number of rice grains for the Sorori Shinzaemon doubling scenario.

    Parameters
    ----------
    n_days : int
        Number of days to calculate (default: 100)

    Returns
    -------
    list_n_grains : list of int
        Grains received per day (length == n_days)
    list_total_grains : list of int
        Cumulative grains up to each day (length == n_days)
    """
    list_n_grains = []
    list_total_grains = []
    total = 0
    grains_today = 1

    for day in range(1, n_days + 1):
        if day == 1:
            grains_today = 1
        else:
            grains_today *= 2
        list_n_grains.append(grains_today)
        total += grains_today
        list_total_grains.append(total)

    return list_n_grains, list_total_grains

# Example - verify with a smaller number of days
# Can attempt a different number
list_n, list_total = compute_sorori_shinzaemon(n_days=8)
print("Day 8 grains:", list_n[-1])
print("Cumulative after 8 days:", list_total[-1])
```

## 7. Estimating Human Consumption
This assignment requires extending the analysis to estimate food support duration.

I did not get a conclusive response to how many grains of rice are in a cup of rice, but for practise purposes let's work with 10,000 grains

The function below provides a general calculation interface.

```python
def estimate_days_of_food(total_grains, grains_per_person_per_day=10000, n_people=1):
    """
    Estimate how many full days a group of people can be supported by the available grains.

    Parameters
    ----------
    total_grains : int
        Total number of grains available
    grains_per_person_per_day : int
        Average grains needed per person per day (default: 10,000 as a placeholder)
    n_people : int
        Number of people consuming the rice (default: 1)

    Returns
    -------
    days : int
        Number of full days the rice can sustain the group
    """
    daily_consumption = grains_per_person_per_day * n_people
    if daily_consumption <= 0:
        raise ValueError("grains_per_person_per_day and n_people must be positive")
    return total_grains // daily_consumption

# Example: estimate support duration from the 100-day total
total_100 = list_total_grains[-1]
print("Grains of rice for 1 person (100 days):", estimate_days_of_food(total_100, grains_per_person_per_day=10000, n_people=1))
print("Grains of rice for 5 people (100 days):", estimate_days_of_food(total_100, grains_per_person_per_day=10000, n_people=5))
```
