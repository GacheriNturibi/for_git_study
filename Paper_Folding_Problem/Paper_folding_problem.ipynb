## 2. Calculating Thickness in Two Ways

### Method A: Using the exponentiation operator (`**`)


```python
# Calculate thickness after one fold
THICKNESS = 0.00008  # meters
folded_thickness = THICKNESS * 2
print("Thickness after 1 fold: {} meters".format(folded_thickness))
```

```python
# Calculate thickness after 43 folds using exponentiation operator (**)
THICKNESS = 0.00008
N_FOLDS = 43
t_43_pow = THICKNESS * (2 ** N_FOLDS)
print("Thickness after 43 folds (meters):", t_43_pow)
```


### Unit Conversion
Convert the thickness into kilometers and round it to two decimal places.

### **Method A**

```python
# Convert to km and format with 2 decimal places
print("Thickness: {:.2f} kilometers".format(t_43_pow / 1000))
```

The distance to the moon is approximately **384,400 km**.  
By comparing the two values, it is clear that the folded paper would really reach the moon.


```python
MOON_KM = 384_400  # km
t_43_km = t_43_pow / 1000
print("Does it reach the moon?", "Yes" if t_43_km >= MOON_KM else "No")
print(f"Paper thickness: {t_43_km:.2f} km / Distance to moon: {MOON_KM:,.0f} km")
```


### **Method B: Using a `for` loop**
Replace the exponential operator with a for loop which will iterate till the 43rd fold.


```python
# Calculate thickness after 43 folds using a for loop
THICKNESS = 0.00008  # meters
N_FOLDS = 43
t_43_loop = THICKNESS
for _ in range(N_FOLDS):
    t_43_loop *= 2
print("Thickness after 43 folds (meters, loop):", t_43_loop)
```

```python
# Check that both methods give the same result
print("Same result?", abs(t_43_loop - t_43_pow) < 1e-12)
```

### Comparing Execution Times
Both methods give the correct result, but how do their execution times compare?


```python
import time # necessary library

THICKNESS = 0.00008
N_FOLDS = 43

start = time.time()
t_43_pow = THICKNESS * (2 ** N_FOLDS)
elapsed_pow = time.time() - start
print("Exponentiation version time: {} [s]".format(elapsed_pow))
```

```python
start = time.time()
t = THICKNESS
for _ in range(N_FOLDS):
    t *= 2
elapsed_loop = time.time() - start
print("Loop version time: {} [s]".format(elapsed_loop))
```

Evidently, using a for loop takes less time, as seen in the outputs.

Exponentiation version time: **7.390975952148438e-05 [s]**

Loop version time: **0.00014591217041015625 [s]**


## 3. Visualizing the Growth with a Graph


```python
# Create list of thickness values (n=0..43)
THICKNESS = 0.00008  # meters
N_FOLDS = 43
thickness_list = [THICKNESS]
current = THICKNESS
for _ in range(N_FOLDS):
    current *= 2
    thickness_list.append(current)

len(thickness_list), thickness_list[:5], thickness_list[-5:]  # preview
```

```python
# Check that there are 44 values (n=0..43)
print("Length OK?", len(thickness_list) == 44)
```

```python
import matplotlib.pyplot as plt

# Basic line plot of thickness
plt.title("Thickness of folded paper")
plt.xlabel("Number of folds")
plt.ylabel("Thickness [m]")
plt.plot(thickness_list)
plt.show()
```

```python
# Customization 1: bigger tick labels and thicker line
plt.title("Thickness of folded paper")
plt.xlabel("Number of folds")
plt.ylabel("Thickness [m]")
plt.tick_params(labelsize=14)
plt.plot(thickness_list, linewidth=2)
plt.show()
```

```python
# Customization 2: dashed line
plt.title("Thickness of folded paper (dashed)")
plt.xlabel("Number of folds")
plt.ylabel("Thickness [m]")
plt.plot(thickness_list, linestyle="--")
plt.show()
```

```python
# Customization 3: markers at some points
plt.title("Thickness of folded paper (markers)")
plt.xlabel("Number of folds")
plt.ylabel("Thickness [m]")
plt.plot(thickness_list, marker="o", markevery=4)
plt.show()
```
