# Dummy Data Creation with NumPy

- Practice graph displays - scatter plots, histograms etc.

## 2. Creating Dummy Data
This assignment requires generating random data from a two-dimensional normal distribution.
The distribution has mean (-3, 0) and covariance matrix:

\[
\begin{bmatrix}
1.0 & 0.8 \\
0.8 & 1.0
\end{bmatrix}
\]

A total of 500 points should be generated with a fixed seed of 0.

```python
import numpy as np

# Fix random seed for reproducibility
np.random.seed(0)

mean0 = [-3, 0]
cov = [[1.0, 0.8], [0.8, 1.0]]
data0 = np.random.multivariate_normal(mean0, cov, 500)
print("Shape of data0:", data0.shape)
data0[:5]
```

## 3. Visualization - Scatter Plot


```python
import matplotlib.pyplot as plt

plt.scatter(data0[:,0], data0[:,1], alpha=0.6, label="Class 0")
plt.xlabel("x1")
plt.ylabel("x2")
plt.title("Scatter plot of generated data")
plt.legend()
plt.show()
```

## 4. Visualization - Histograms
Use the same x-axis limits for comparison.


```python
fig, axes = plt.subplots(1, 2, figsize=(10,4)) # visualize on same row using subplots

axes[0].hist(data0[:,0], bins=30, alpha=0.7, color="blue")
axes[0].set_title("Histogram of x1")
axes[0].set_xlim(-7, 5)

axes[1].hist(data0[:,1], bins=30, alpha=0.7, color="green") # changed colors for comparison
axes[1].set_title("Histogram of x2")
axes[1].set_xlim(-7, 5)

plt.show()
```

## 5. Adding Another Distribution
This assignment requires generating another 500 points from a distribution with mean (0, -3) and the same covariance.
Then plot both datasets together with a legend.

```python
mean1 = [0, -3]
data1 = np.random.multivariate_normal(mean1, cov, 500)
print("Shape of data1:", data1.shape)

plt.scatter(data0[:,0], data0[:,1], alpha=0.6, label="Class 0")
plt.scatter(data1[:,0], data1[:,1], alpha=0.6, label="Class 1")
plt.xlabel("x1")
plt.ylabel("x2")
plt.title("Scatter plot of data0 and data1")
plt.legend()
plt.show()
```

## 6. Data Combination
Combining the two ndarrays into a single dataset with shape (1000, 2).

```python
data_combined = np.vstack([data0, data1])
print("Combined data shape:", data_combined.shape)
```

## 7. Labeling the Data
Add a label column (0 for the first dataset, 1 for the second).
The final dataset should have shape (1000, 3).

```python
labels0 = np.zeros((500,1), dtype=int)
labels1 = np.ones((500,1), dtype=int)
labels = np.vstack([labels0, labels1])

dataset = np.hstack([data_combined, labels])
print("Final dataset shape:", dataset.shape)
dataset[:5]
```
